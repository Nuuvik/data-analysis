## Dataset Description

The dataset comprises information about customers and their purchasing behavior.

### Columns:
- **CustomerID**: Unique identifier for each customer.
- **Genre**: Gender of the customer.
- **Age**: Age of the customer.
- **Annual Income (k$)**: Annual income of the customer in thousands of dollars.
- **Spending Score (1-100)**: Score assigned to the customer based on their purchasing behavior and financial status, ranging from 1 to 100.

This dataset can be utilized for various analyses such as customer segmentation, market targeting, and understanding consumer behavior patterns.

### Link: https://www.kaggle.com/datasets/shwetabh123/mall-customers/data


```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
print("\nАтрибут Genre")

df['Genre'].value_counts().plot(kind='bar', edgecolor='black')
plt.title('Distribution of Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

Как видно по гистограмме - посетителей женского пола примерно на 20 больше, чем мужского

```python
print("\nАтрибут Age")

print("\nСредние значения:")
print(df['Age'].mean())

print("\nСтандартное отклонение:")
print(df['Age'].std())

print("\nГистограмма распеделения значений:")
plt.hist(df['Age'], bins=50, edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

missing_values_age = df['Age'].isnull().sum()
print(f"Пропущенные значения в 'Age': {missing_values_age}")
print("\nВ случае если будут пропущенные значения, можно использовать медиану")
```

Как видно по гистограмме - выбросы происходят на отметке 19 и ~33 года

```python
print("\nАтрибут Annual Income (k$)")

print("\nСредние значения:")
print(df['Annual Income (k$)'].mean())

print("\nСтандартное отклонение:")
print(df['Annual Income (k$)'].std())

print("\nГистограмма распеделения значений:")
plt.hist(df['Annual Income (k$)'], bins=50, edgecolor='black')
plt.title('Distribution of Annual Income (k$)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

missing_values_anual = df['Annual Income (k$)'].isnull().sum()
print(f"Пропущенные значения в 'Annual Income (k$)': {missing_values_anual}")
print("\nВ случае если будут пропущенные значения, можно использовать медиану")
```

Как видно по гистограмме - выбросы происходят на отметке ~77k$

```python
print("\nSpending Score (1-100)")

print("\nСредние значения:")
print(df['Spending Score (1-100)'].mean())

print("\nСтандартное отклонение:")
print(df['Spending Score (1-100)'].std())

print("\nГистограмма распеделения значений:")
plt.hist(df['Spending Score (1-100)'], bins=50, edgecolor='black')
plt.title('Distribution of Spending Score (1-100)')
plt.xlabel('Spending Score (1-100)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

missing_values_spend = df['Spending Score (1-100)'].isnull().sum()
print(f"Пропущенные значения в 'Spending Score (1-100)': {missing_values_spend}")
print("\nВ случае если будут пропущенные значения, можно использовать медиану")
```

Как видно по гистограмме - выбросы происходят на отметке ~41 spending score
